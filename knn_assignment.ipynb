```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haberman = pd.read_csv("haberman.data")
```

```python
haberman.columns = ["age", "year of operation", "number of auxillary nodes", "survival status"]
```

```python
haberman.isnull().sum()
```

```python
haberman.head(3)
```

```python
sns.pairplot(haberman, hue="survival status")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = haberman[["age", "number of auxillary nodes"]]
```

```python
y = haberman["survival status"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=41)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
